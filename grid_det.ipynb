```python
dots = cv2.imread('dutti.png')
dots = cv2.cvtColor(dots,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(dots)
```

```python
blurred_image = cv2.GaussianBlur(dots, (5, 5), 0)
```

```python
# found ,corners = cv2.findCirclesGrid(dots,(7,7),cv2.CALIB_CB_SYMMETRIC_GRID)
found, corners = cv2.findCirclesGrid(blurred_image, (7,7), cv2.CALIB_CB_SYMMETRIC_GRID)
```

```python
cv2.drawChessboardCorners(dots,(7,7),corners,found)
```
