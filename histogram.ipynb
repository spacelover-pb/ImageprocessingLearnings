# histogram equalization

```python

puppy =  cv2.imread('puppy.jpeg',0)
```

```python
def display(img,cmap = None):
    fig = plt.figure(figsize =(10,8))
    ax =  fig.add_subplot(111)
    ax.imshow(img,cmap)
```

```python
display(puppy,cmap ='gray')
```

```python
puppy.shape
```

```python
hist_values = cv2.calcHist([puppy],channels =[0],mask = None,histSize =[256],ranges = [0, 256])
```

```python
plt.plot(hist_values)
```

```python
eq_puppy = cv2.equalizeHist(puppy)
```

```python
display(eq_puppy,cmap = 'gray')
```

```python
hist_values = cv2.calcHist([eq_puppy],channels =[0],mask = None,histSize =[256],ranges = [0, 256])
```

```python
plt.plot(hist_values)
```
