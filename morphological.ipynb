```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Function to load image
def load_img():
    blank_img = np.zeros((600,600))
    font = cv2.FONT_HERSHEY_SIMPLEX
    cv2.putText(blank_img,text = 'ABCDE',org =(50,300),fontFace = font,fontScale = 5,color = (255,255,255), thickness = 10)
    return blank_img
```

```python
# Function to display image
def display_img(img):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap = 'gray')
```

```python
img = load_img()
display_img(img)
```

```python
kernel = np.ones((5,5),dtype=np.uint8)
kernel
```

```python
result = cv2.erode(img,kernel,iterations = 2)
```

```python
display_img(result)
```


# sober operator

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Function to display image
def display_img(img):
    fig = plt.figure(figsize=(12, 10))
    ax = fig.add_subplot(111)
    ax.imshow(img,cmap = 'gray')
```

```python
img = cv2.imread('Sudoku.jpg',0)
display_img(img)
```

```python
sobel_x =  cv2.Sobel(img,cv2.CV_64F,1,0,ksize = 5)
```

```python
display_img(sobel_x)
```

```python
sobel_y =  cv2.Sobel(img,cv2.CV_64F,0,1,ksize = 5)
```

```python
display_img(sobel_y)
```

```python
laplacian =  cv2.Laplacian(img,cv2.CV_64F)
```

```python
display_img(laplacian)
```

```python
blended =  cv2.addWeighted(src1 = sobel_x,alpha = 0.5,src2 = sobel_y,beta = 0.5,gamma = 0)
```

```python
display_img(blended)
```

```python
ret,th1 = cv2.threshold(img,100,255,cv2.THRESH_BINARY)
```

```python
display_img(th1)
```

```python
kernel = np.ones((4,4),np.uint8)
```

```python
gradient = cv2.morphologyEx(blended, cv2.MORPH_GRADIENT, kernel)
```

```python
display_img(gradient)
```
