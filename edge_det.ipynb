```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
img = cv2.imread('dog_face.jpg')
# img = cv2.cvtColor(img,cv2.COLOR_BGR2RGB)
```

```python
plt.imshow(img)
```

```python
edges = cv2.Canny(image = img,threshold1 = 127,threshold2 = 127)
```

```python
plt.imshow(edges)
```

```python
edges = cv2.Canny(image = img,threshold1 = 0,threshold2 = 255)

plt.imshow(edges)
```

```python
med_val = np.median(img)
```

```python
med_val
```

```python
lower = int(max(0,0.7*med_val))   # lower threshold either 0 or 70% of the median value whichever is greater
upper = int(min(255,1.3*med_val))   # upper threshold to either 130% of the median or max 255 whichever is smaller
```

```python
edges = cv2.Canny(image = img,threshold1 = lower,threshold2 = upper)
plt.imshow(edges)
```

```python
blurred_img = cv2.blur(img,ksize = (5,5))
```

```python
edges = cv2.Canny(image = blurred_img,threshold1 = lower,threshold2 = upper)
plt.imshow(edges)
```
